```python
import pandas as pd
import warnings 
from pprint import pprint
warnings.filterwarnings(action='ignore')
```

```python
df = pd.read_csv('result.csv')
df.head()
```

```python
skid=[]
sales=[]
grosmargin=[]
spend1=[]
spend2=[]

def get_sumup(df):
      sku_ids = list(df.SKU_ID.unique())
      for id in sku_ids:
            skuvalues = df[df.SKU_ID == id]
            sales_sumup = skuvalues['Sales'].sum()
            gross_margin_count = len(skuvalues['Gross Margin'])
            spend1_sumup = skuvalues['Spend 1'].sum()
            spend2_count = len(skuvalues['Spend 1'])

            skid.append(id)
            sales.append(sales_sumup)
            grosmargin.append(gross_margin_count)
            spend1.append(spend1_sumup)
            spend2.append(spend2_count)
            

get_sumup()
print(skid, sales,grosmargin,spend1, spend2) 

data = pd.DataFrame({ 
      "SKU_ID":skid, 
      'Sales': sales, 
      "Gross Margin":grosmargin,
      "Spend 1": spend1,
      "Spend 2" : spend2
})
data
```
